```python
# Import libraries
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr
from PIL import Image

# Load the pre-trained Stable Diffusion model
device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id).to(device)
```

```python
# Define the function to generate images based on the prompt and additional parameters
def generate_image(character_concept, avoid_terms, creativity_level, detail_level):
    with torch.no_grad():
        # Generate an image with the Stable Diffusion model
        images = pipe(
            prompt=character_concept,
            negative_prompt=avoid_terms if avoid_terms else None,
            guidance_scale=creativity_level,
            num_inference_steps=detail_level
        ).images
    return images[0]  # Return the first generated image
```

```python
from diffusers import StableDiffusionPipeline
import gradio as gr

# Load the Stable Diffusion model
model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
model.to("cuda")  # Use GPU for faster generation
pipe.enable_sequential_cpu_offload()

# Function to generate character design with customized parameters
def generate_character_design(character_concept, creativity_level=7.5, detail_level=50):
    """
    Generate a character design based on the character concept with customization options:
    - creativity_level: Controls adherence to prompt (higher values = more aligned with prompt)
    - detail_level: Controls quality and detail (higher values increase quality but slow down generation)
    """

    # Define negative prompts to avoid overly realistic elements and maintain a stylized look
    negative_prompt = "photorealistic, photography, realism, background, multiple faces "

    # Character design styles to include in the prompt
    character_styles = "character design,anime style,3D portrait, fantasy, defined facial features, video game characters"

    # Combine the prompt with character styles
    combined_prompt = f"{character_concept}, {character_styles}"

    # Generate the character image
    image = model(
        prompt=combined_prompt,
        negative_prompt=negative_prompt,
        guidance_scale=creativity_level,
        num_inference_steps=detail_level
    ).images[0]

    return image

# Set up Gradio Interface with input parameters
interface = gr.Interface(
    fn=generate_character_design,
    inputs=[
        gr.Textbox(label="Character Concept", placeholder="Describe your character"),
        gr.Slider(label="Creativity Level", minimum=5, maximum=20, step=0.5, value=7.5),
        gr.Slider(label="Detail Level", minimum=10, maximum=100, step=5, value=50)
    ],
    outputs="image",
    title="AI Character Design Tool",
    description="Generate unique character designs with custom settings for creativity and detail."
)

interface.launch()
```
